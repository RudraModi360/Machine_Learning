```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\rudra\\Downloads\\Melbourne_housing_FULL.csv")
```

```python
dt=df.drop(["Address","Type","Method","SellerG","Date","YearBuilt"],axis="columns")
dt.isna().sum()
```

```python
dt
```

```python
dt["Postcode"]=dt["Postcode"].fillna(dt["Postcode"].mean())
```

```python
X_train,X_test,y_train,y_test=train_test_split(dt[["Rooms","Distance","Postcode","Bedroom2","Bathroom","Car","Landsize","BuildingArea","CouncilArea","Lattitude","Longtitude","Regionname","Propertycount"]],dt.Price)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
dt["Regionname"]=le.fit_transform(dt.Regionname)
```

```python
dt
```

```python
dt.isna().sum()
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
print(model.score(X_train,y_train))
model.score(X_test,y_test)
```

```python
from sklearn.linear_model import Lasso
ls=Lasso()
ls.fit(X_train,y_train)
print(ls.score(X_train,y_train))
ls.score(X_test,y_test)
```

```python
from sklearn.linear_model import Ridge
rd=Ridge()
rd.fit(X_train,y_train)
print(rd.score(X_train,y_train))
rd.score(X_test,y_test)
```
