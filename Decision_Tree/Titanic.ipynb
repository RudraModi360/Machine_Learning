```python
import pandas as pd
from sklearn import tree
```

```python
df=pd.read_csv("C:\\Users\\rudra\\Downloads\\titanic.csv")
df
```

```python
inputs=df.drop("Survived",axis="columns")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
inputs["PassengerId"]=le.fit_transform(inputs["PassengerId"])
inputs["Pclass"]=le.fit_transform(inputs["Pclass"])
inputs["Name"]=le.fit_transform(inputs["Name"])
inputs["Sex"]=le.fit_transform(inputs["Sex"])
inputs["Age"]=le.fit_transform(inputs["Age"])
inputs["SibSp"]=le.fit_transform(inputs["SibSp"])
inputs["Parch"]=le.fit_transform(inputs["Parch"])
inputs["Ticket"]=le.fit_transform(inputs["Ticket"])
inputs["Fare"]=le.fit_transform(inputs["Fare"])
inputs["Cabin"]=le.fit_transform(inputs["Cabin"])
inputs["Embarked"]=le.fit_transform(inputs["Embarked"])
inputs
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(inputs,df["Survived"])
```

```python
model=tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
