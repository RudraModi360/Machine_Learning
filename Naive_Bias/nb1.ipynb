```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
```

```python
df=pd.read_csv("C:\\Users\\rudra\\Downloads\\titanic.csv")
```

```python
df.head(5)
```

```python
inputs=df.drop(["Name","PassengerId","Ticket","Cabin","Embarked"],axis="columns")
inputs
```

```python
survived=df["Survived"]
inputs.drop("Survived",axis="columns",inplace=True)
inputs
```

```python
gender=pd.get_dummies(inputs["Sex"])
inputs=pd.concat([gender,inputs],axis="columns")
inputs.drop("Sex",axis="columns",inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
```

```python
inputs["Sex"]=le.fit_transform(inputs["Sex"])
```

```python
inputs["Age"]=inputs.Age.fillna(int(inputs.Age.mean()))
inputs["Age"][888]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(inputs,survived)
```

```python
nb=GaussianNB()
nb.fit(X_train,y_train)
nb.score(X_test,y_test)
```

```python
nb.predict_proba(X_test)
```
