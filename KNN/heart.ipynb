```python
from sklearn.decomposition import PCA
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("C:\\Users\\rudra\\Downloads\\heart.csv")
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df["ExerciseAngina"]=le.fit_transform(df["ExerciseAngina"])
df["ST_Slope"]=le.fit_transform(df["ST_Slope"])
df["RestingECG"]=le.fit_transform(df["RestingECG"])
df["ChestPainType"]=le.fit_transform(df["ChestPainType"])
df["Sex"]=le.fit_transform(df["Sex"])
df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
HeartDisease=df.HeartDisease
data=df.drop(["HeartDisease"],axis="columns")
```

```python
X_train,X_test,y_train,y_test=train_test_split(data,HeartDisease)
```

```python
model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.preprocessing import StandardScaler

ss=StandardScaler()
X_scaled=ss.fit_transform(X_test)
```

```python
pca=PCA(0.95)
X_pca=pca.fit_transform(X_train)
X_pca_test=pca.fit_transform(X_test)
```

```python
model.fit(X_pca,y_train)
model.score(X_pca_test,y_test)
```
