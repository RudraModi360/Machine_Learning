```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv("C:\\Users\\rudra\\Downloads\\homeprices.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged=pd.concat([df,dummies],axis="columns")
merged
```

```python
df_dummies=merged.drop(["west windsor","town"],axis="columns")
df_dummies
```

```python
model=linear_model.LinearRegression()
model.fit(df_dummies[["area","monroe township","robinsville"]],df_dummies.price)
```

```python
model.predict([[3400,True,False]])
```
